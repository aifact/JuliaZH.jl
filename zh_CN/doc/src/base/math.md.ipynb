## 自定义二元运算符

某些 unicode 字符可用于定义新的支持中缀表示法的二元运算符。
例如，
```⊗(x,y) = kron(x,y)```
定义 `⊗` (otimes) 为 Kronecker 积，
并且可以通过中缀语法将它作为一个二元运算符调用：
```C = A ⊗ B```
也可以使用常用的前缀语法
```C = ⊗(A,B)```。

其他支持这种扩展的字符包括
\odot `⊙`
和
\oplus `⊕`

The complete list is in the parser code:
<https://github.com/JuliaLang/julia/blob/master/src/julia-parser.scm>

像 `*` 一样解析的包括（按优先级排列）
`* / ÷ % & ⋅ ∘ × |\\| ∩ ∧ ⊗ ⊘ ⊙ ⊚ ⊛ ⊠ ⊡ ⊓ ∗ ∙ ∤ ⅋ ≀ ⊼ ⋄ ⋆ ⋇ ⋉ ⋊ ⋋ ⋌ ⋏ ⋒ ⟑ ⦸ ⦼ ⦾ ⦿ ⧶ ⧷ ⨇ ⨰ ⨱ ⨲ ⨳ ⨴ ⨵ ⨶ ⨷ ⨸ ⨻ ⨼ ⨽ ⩀ ⩃ ⩄ ⩋ ⩍ ⩎ ⩑ ⩓ ⩕ ⩘ ⩚ ⩜ ⩞ ⩟ ⩠ ⫛ ⊍ ▷ ⨝ ⟕ ⟖ ⟗`
